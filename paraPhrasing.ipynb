```python
from simpletransformers.t5 import T5Model
from sklearn.model_selection import train_test_split
import sklearn
```

```python
dataset_df.columns = ["input_text","target_text"]
dataset_df["prefix"] = "paraphrase"
```

```python
train_data,test_data = train_test_split(dataset_df,test_size=0.1)
```

```python
args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "max_seq_length": 256,
    "num_train_epochs": 4,
    # "num_beams": None,
    "do_sample": True,
    "top_k": 50,
    "top_p": 0.95,
    "use_multiprocessing": False,
    "save_steps": -1,
    "save_eval_checkpoints": True,
    "evaluate_during_training": False,
    'adam_epsilon': 1e-08,
    'eval_batch_size': 6,
    # 'use_cuda':False,
    'fp16': False,
    'gradient_accumulation_steps': 16,
    'learning_rate': 0.0003,
    'max_grad_norm': 1.0,
    'n_gpu': 1,
    'seed': 42,
    'train_batch_size': 6,
    'warmup_steps': 0,
    'weight_decay': 0.0
}
```

```python
model = T5Model("t5","t5-small", args=args,use_cuda=True)
```

```python
model.train_model(train_data, eval_data=test_data, use_cuda=True,acc=sklearn.metrics.accuracy_score)
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

device = "cuda"
model_id = "t5-small"
model = T5ForConditionalGeneration.from_pretrained(model_id).to(device)
# tokenizer = T5Tokenizer.from_pretrained(model_id)
# from transformers import T5Tokenizer

tokenizer = T5Tokenizer.from_pretrained("t5-small")



```

```python
# import pandas as pd 
# df = pd.read_csv("/kaggle/input/outputs/output.csv")
# vocab = df.iloc[0].values.tolist()
# vocab[:3]
# import pandas as pd 
df = pd.read_csv('/content/paraphrased_output3.csv')
# df = pd.read_csv("drive/MyDrive/Colab Notebooks/paraphrased_output.csv")
# vocab = df.iloc[0].values.tolist()
vocab = df.iloc[:, 0].values.tolist()
# vocab[:3]
```

```python
vocab = vocab[-5:]
# vocab
```

```python
max([len(i.split()) for i in vocab])
```

```python
encodings = tokenizer("\n\n".join(vocab), return_tensors="pt")
```

```python
encodings
```

```python
import torch
from tqdm import tqdm

max_length = model.config.n_positions
stride =  5
seq_len = encodings.input_ids.size(1)

nlls = []
prev_end_loc = 0
for begin_loc in tqdm(range(0, seq_len, stride)):
    end_loc = min(begin_loc + max_length, seq_len)
    trg_len = end_loc - prev_end_loc  # may be different from stride on last loop
    input_ids = encodings.input_ids[:, begin_loc:end_loc].to(device)
    target_ids = input_ids.clone()
    target_ids[:, :-trg_len] = -100

    with torch.no_grad():
        outputs = model(input_ids, labels=target_ids)

        # loss is calculated using CrossEntropyLoss which averages over input tokens.
        # Multiply it with trg_len to get the summation instead of average.
        # We will take average over all the tokens to get the true average
        # in the last step of this example.
        neg_log_likelihood = outputs.loss * trg_len

    nlls.append(neg_log_likelihood)

    prev_end_loc = end_loc
    if end_loc == seq_len:
        break

ppl = torch.exp(torch.stack(nlls).sum() / end_loc)
```
