```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
# device = "cuda"
# Load the pre-trained GPT-2 model and tokenizer
model_id = "gpt2-medium"
model = GPT2LMHeadModel.from_pretrained(model_id).to(device)
tokenizer = GPT2Tokenizer.from_pretrained(model_id)
```

```python
def get_log_likelihood(text):

  # Tokenize the input text and generate model input
  input_ids = tokenizer.encode(text, return_tensors='pt')

  # Generate model output and compute log-likelihood
  output = model(input_ids=input_ids, labels=input_ids)
  log_likelihood = output.loss.item()

  # print("Log-likelihood:", log_likelihood)
  return log_likelihood

df["log_likelihood_org"] = df.org_sentences.apply(lambda x: get_log_likelihood(x))
df["log_likelihood_para"] = df.paraphrased.apply(lambda x: get_log_likelihood(x))
```

```python
df.head()
```

```python
df['diff_log_likelihood'] = abs(df['log_likelihood_org'] - df['log_likelihood_para'])
```
